```python
## What are the types of the planets by their radius?
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df[df["koi_prad"] < 20]["koi_prad"],
    bins=30
)

plt.title("Planet Radius Distribution (<20 Earth Radii)")
plt.xlabel("Planet Radius (Earth Radii)")
plt.ylabel("Count")

plt.show()
```

```python
def classify_planet(radius):
    if radius < 1.25:
        return "Earth-like"
    elif radius < 2:
        return "Super Earth"
    elif radius < 6:
        return "Neptune-like"
    else:
        return "Gas Giant"

df["planet_type"] = df["koi_prad"].apply(classify_planet)
```

```python
sns.countplot(data=df, x= "planet_type")

plt.xticks(rotation=20)
plt.xlabel("Planet types")
plt.show()
```
